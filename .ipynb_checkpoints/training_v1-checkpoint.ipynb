```python
config = particles.ParticlesConfig()
ROOT_DIR = os.getcwd()
PARTICLE_DIR = 'C:/Users/Sohrab/Documents/crack/EILNet/images/subset_1/data/'
COCO_MODEL_PATH = 'C:/Users/Sohrab/Documents/crack/EILNet/mask_rcnn_coco.h5'
class InferenceConfig(config.__class__):
    GPU_COUNT = 1
    IMAGES_PER_GPU =1
    IMAGE_RESIZE_MODE = 'square'
    IMAGE_MIN_DIM = 768
    IMAGE_MAX_DIM = 1024
    LEARNING_RATE = 0.001
    LEARNING_MOMENTUM = 0.9
config_inference = InferenceConfig()

DEVICE = '/GPU:0'
TEST_MODE = 'training'
def get_ax(rows=1, cols=1, size=16):
    _, ax = plt.subplots(rows,cols,figsize=(size*cols,size*rows))
    return ax
```

```python
train_data = particles.ParticlesDataset()
train_data.load_particles(PARTICLE_DIR,'annotations.json','train')
train_data.prepare()
val_data = particles.ParticlesDataset()
val_data.load_particles(PARTICLE_DIR,'annotations.json','val')
val_data.prepare()
image_id = np.random.choice(range(0,10))
image = train_data.load_image(image_id)
mask,class_ids=train_data.load_mask(image_id)
visualize.display_top_masks(image, mask, class_ids, train_data.class_names)
```

```python
with tf.device(DEVICE):
    #changelogs
    model = modellib.MaskRCNN(mode='training', config = config, model_dir = 'C:/Users/Sohrab/Documents/crack/code/libnet/LIBNet/logs')
init_with = 'coco'
if init_with == 'imagenet':
    model.load_weights(model.get_imagenet_weights(),by_name=True)
elif init_with == 'coco':
    model.load_weights(COCO_MODEL_PATH, by_name=True,exclude=['mrcnn_class_logits','mrcnn_bbox_fc','mrcnn_bbox','mrcnn_mask'])
elif init_with=='last':
    model.load_weights(model.find_last(),by_name=True)
    
```
